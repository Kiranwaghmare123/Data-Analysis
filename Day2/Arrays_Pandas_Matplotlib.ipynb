```python
#Visualization
#Matplotlib: A comprehensive library for creating static, animated and interactive visualizations in python
#Seaborn: Built on top of matplotlib, it provides a high-level interface for drawing attractive and
#informative statistical graphics
#Instalation: pip install mayplotlib
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4,5]
Y = [2,3,5,7,9]

plt.plot(X,Y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Plot cutomization
# Line Style and color
plt.plot(X,Y,color='red',linestyle='--',marker='o')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show
```

```python
#Add grid
plt.plot(X,Y,color='red',linestyle='--',marker='o')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
#Legends
plt.plot(X,Y, label='Prime number')
plt.legend()
plt.show()
```

```python
#Subplots
plt.subplot(2,1,1) #row, col, panel number
plt.plot(X,Y,'r-')
plt.title('Linear Plot')

plt.subplot(2,1,2)
plt.plot(X,[1,4,9,16,25],'b-')
plt.title('Square Plot')
plt.show()
```

```python
#Example:Combining multiple plots
import numpy as np

t = np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.legend(['Linear','Square','Cubic'])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Combined Plots')
plt.show()
```
