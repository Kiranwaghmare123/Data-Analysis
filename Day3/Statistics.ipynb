```python
import numpy as np

data =[10,20,30,40,50]
mean = np.mean(data)
mean
```

```python
median = np.median(data)
median
```

```python
from scipy import stats
mode = stats.mode(data)
mode
```

```python
variance = np.var(data)
variance
```

```python
std_dev = np.std(data)
std_dev
```

```python
range_val = np.ptp(data)
iqr = np.percentile(data,75) - np.percentile(data,25)
print(iqr)
range_val
```

```python
#Normal distribution
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(data)
plt.show()
```

```python
dat_norml = np.random.normal(0,1,1000)
sns.histplot(dat_norml)
plt.title("Normal Distribution")
#plt.show()
```

```python
# Uniform distribution
data_uniform = np.random.uniform(0,1,1000)
sns.histplot(data_uniform)
plt.title("Uniform Distribution")
```

```python
#Binomial distribution
data_binomial = np.random.binomial(10,0.5,1000)
sns.histplot(data_binomial, kde=True)
plt.title("Binomial Distribution")
```

```python
#Hypothesis testing
#p-test, t-test
group1 = np.random.normal(10, 5,100)
group2 = np.random.normal(15, 5,100)
t_stat,p_stat = stats.ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("P-value:", p_stat)
```

```python
#Chi-square test
observed = np.array([[10,20,30],[40,50,60]])
chi2, p , dof, ex = stats.chi2_contingency(observed)
print("Chi-square statistic:", chi2)
print("P-value:", p)
```

```python
data1 = np.random.rand(100)
data2 = np.random.rand(100)
correlation, p_value = stats.pearsonr(data1, data2)
print("Correlation coefficient:", correlation)
print("P-value:", p_value)
```
